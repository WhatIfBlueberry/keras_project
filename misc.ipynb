```python
import tensorflow as tf

model = tf.keras.models.load_model('blitzinator.keras') 
```

```python
import keras
import matplotlib.pyplot as plt
```

```python
# Preprocess the image
img = tf.keras.utils.load_img('catsndogs//PetImages//Cat//10272.jpg', target_size=(180, 180))
img_array = tf.keras.utils.img_to_array(img)
img_array = tf.expand_dims(img_array, 0)  # Create batch axis
img_array = img_array / 255.0  # Normalize pixel values

# Make prediction
predictions = model.predict(img_array)
```

```python
img = keras.utils.load_img("catsndogs//PetImages//Cat//10272.jpg", target_size=(180, 180))
plt.imshow(img)

img_array = keras.utils.img_to_array(img)
img_array = keras.ops.expand_dims(img_array, 0)  # Create batch axis

predictions = model.predict(img_array)
score = float(keras.ops.sigmoid(predictions[0][0]))
print(f"This image is {100 * (1 - score):.2f}% cat and {100 * score:.2f}% dog.")
```
